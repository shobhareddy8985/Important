```python
import pandas as pd
import matplotlib.pyplot as mplot
# inline function will display the graph after running the program
%matplotlib inline 
train=pd.read_csv('train.csv')
#train.tail()
result =train.groupby(["YearBuilt"])["MSSubClass"].count()
mplot.ylabel("count(*)")
mplot.xlabel("year")
dtype='result'
print(result)
mplot.plot(result,color='green',marker='^',markeredgecolor='red')
```
